```python

variable_names = ["erythema"
        , "scaling"
        , "definite borders"
        , "itching"
        , "koebner phenomenon"
        , "polygonal papules"
        , "follicular papules"
        , "oral mucosal involvement"
        , "knee and elbow involvement"
        , "scalp involvement"
        , "family history"
        , "melanin incontinence"
        , "eosinophils in the infiltrate"
        , "PNL infiltrate"
        , "fibrosis of the papillary dermis"
        , "exocytosis"
        , "acanthosis"
        , "hyperkeratosis"
        , "parakeratosis"
        , "clubbing of the rete ridges"
        , "elongation of the rete ridges"
        , "thinning of the suprapapillary epidermis"
        , "spongiform pustule"
        , "munro microabcess"
        , "focal hypergranulosis"
        , "disappearance of the granular layer"
        , "vacuolisation and damage of basal layer"
        , "spongiosis"
        ,"saw-tooth appearance of retes"
        ,"follicular horn plug"
        ,"perifollicular parakeratosis"
        ,"inflammatory monoluclear inflitrate"
        ,"band-like infiltrate"
        ,"Age"]

pathology_names = ("psoriasis", "seboreic dermatitis", "lichen planus", "pityriasis rosea", "cronic dermatitis", "pityriasis rubra pilaris") 
```

**División de sets de datos**

```python
from sklearn.model_selection import train_test_split

print("Total de muestras: " + str(len(data_targets)))

X_train, X_test, y_train, y_test = train_test_split(data_variables, data_targets, test_size=0.2, random_state=4) 
#Investigá: qué utilidad tiene aclarar el random_state?

print("Muestras de entrenamiento: " + str(len(y_train)))

print("Muestras de prueba: " + str(len(y_test)))

```

**Naive Bayes gaussiano**

```python

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
model = gnb.fit(X_train, y_train)
```

El Naive Bayes gaussiano asume que las features tienen una distribución gaussiana. El modelo se ajusta por maximum likelihood hasta encontrar los mejores parámetros (media y desvío estándar) para describir la distribución de cada feature en cada clase. 
-  Graficá la distribución de la picazón (itching) para las 6 patologías del modelo generativo. (Pista: mirar los atributos 'theta_' y 'sigma_' de un objeto GaussianNB.


```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
#"itching" es index 3
for pathology in range(6):
    sigma_itching = math.sqrt(gnb.sigma_[pathology][3])
    mu_itching = gnb.theta_[pathology][3]
    x = np.linspace(-5, 5, 100) #Vector horizontal para plot
    plt.plot(x, stats.norm.pdf(x,mu_itching, sigma_itching))
    plt.title(pathology_names[pathology])
    plt.show()
```

-  Graficá la distribución de la edad para las 6 patologías del modelo generativo. 


```python
for pathology in range(6):
    sigma_edad = math.sqrt(gnb.sigma_[pathology][-1])
    mu_edad = gnb.theta_[pathology][-1]
    x = np.linspace(0, 100, 100) #Vector horizontal para plot
    plt.plot(x, stats.norm.pdf(x,mu_edad, sigma_edad))
    plt.title(pathology_names[pathology])
    plt.show()
```

```python
predictions = model.predict(X_test)
```

```python
porc=sum(predictions==y_test)/y_test.shape[0]*100
print("El porcentaje de lesiones clasificados correctamente es: {}%".format(porc))

#Probar: qué pasa si aumento el tamaño del set de testeo y disminuyo el de entrenamiento? Y viceversa?
```

**Búsqueda de hiperpárametros** 

La constante de smoothing se define como una proporción de la mayor varianza encontrada en la distribución de features, y se sumará a todas las varianzas para estabilizar el proceso de ajuste. Calcular el porcentaje de lesiones clasificadas correctamente para los siguientes valores de smoothing: 

-  1e-3
-  1e-5
-  1e-7
-  1e-9
-  1e-11
-  1e-13


```python
gnb = GaussianNB(var_smoothing=1e-1)
model = gnb.fit(X_train, y_train)
predictions = model.predict(X_test)
porc=sum(predictions==y_test)/y_test.shape[0]*100
print("El porcentaje de lesiones clasificados correctamente es: {}%".format(porc))
```

Los factores del dataset son tanto histopatológicos (requieren de una biopsia) como clínicos. ¿Qué pasa si tomamos sólo los factores clínicos? ¿El mismo var_smoothing sigue siendo el más óptimo?

```python
indices=list(range(11))
indices=indices + [-1]
data_variables_clinical = np.take(data_variables,indices,axis=1)
```

```python
X_train_clinical, X_test_clinical, y_train_clinical, y_test_clinical = train_test_split(data_variables_clinical, data_targets, test_size=0.2, random_state=4)
gnb_clinical = GaussianNB(var_smoothing=1e-4)
model = gnb_clinical.fit(X_train_clinical, y_train_clinical)
predictions = model.predict(X_test_clinical)
porc=sum(predictions==y_test_clinical)/y_test_clinical.shape[0]*100
print("El porcentaje de lesiones clasificados correctamente es: {}%".format(porc))
```
